### Read csv file and store into pandas dataframe

```python
hosp_gen_info_df = pd.read_csv("Raw-data-Hospital_General_Information.csv")
hosp_gen_info_df.head()
```

```python
# View Directory
#dir()
```

### View the columns in the dataframe and work on removing unnesessary columns

```python
hosp_gen_info_df.columns
```

### Drop unwanted columns

```python

cleaned_hosp_gen_info_df = hosp_gen_info_df[['Facility ID', 'Facility Name', 'Address', 'City', 'State', 'ZIP Code',
                                           'County Name', 'Phone Number', 'Hospital Type', 'Hospital Ownership',
                                           'Emergency Services','Meets criteria for promoting interoperability of EHRs',  
                                             'Hospital overall rating', 'Hospital overall rating footnote']]

cleaned_hosp_gen_info_df  = cleaned_hosp_gen_info_df .dropna()
```

### Rename remaining columns

```python
cleaned_hosp_gen_info_df=cleaned_hosp_gen_info_df.rename(columns={'Facility ID':'facility_id', 'Facility Name':'facility_name', 
                                                                  'Address':'address', 'City':'city', 'State':'state',
                                                                  'ZIP Code':'zip_code','County Name':'county_name', 
                                                                  'Phone Number':'phone_no', 'Hospital Type':'hosp_type',
                                                                  'Hospital Ownership':'hosp_owner','Emergency Services':'emergency_srvc',
                                                                  'Meets criteria for promoting interoperability of EHRs':'ehr_compartibility',  
                                                                  'Hospital overall rating':'hosp_overall_rating', 
                                                                  'Hospital overall rating footnote':'hosp_overall_rating_footnote'})
    
    
        
cleaned_hosp_gen_info_df.head()
```

### View the newly named columns

```python
cleaned_hosp_gen_info_df.columns
```

### Repeat the same cleanup process with the second csv file

```python
#Read the csv and store into pandas dataframe


state_dud_df = pd.read_csv("RawData-State_Drug_Utilization-Data.csv")
state_dud_df.head(10)
```

```python
#View columns


state_dud_df.columns
```

```python
# Rename columns


cleaned_state_dud_df=state_dud_df[['Utilization Type', 'State', 'Labeler Code', 'Product Code',
       'Package Size', 'Year', 'Quarter', 'Product Name', 'Units Reimbursed',
       'Suppression Used', 'Number of Prescriptions',
       'Total Amount Reimbursed', 'Medicaid Amount Reimbursed',
       'Non Medicaid Amount Reimbursed', 'Quarter Begin', 'Quarter Begin Date',
       'Latitude', 'Longitude', 'Location', 'NDC']]
cleaned_state_dud_df=cleaned_state_dud_df.rename(columns={'Utilization Type':'utilization_type', 'State':'state', 
                                                          'Labeler Code':'labeler_code', 'Product Code':'product_code',
                                                          'Package Size':'package_size', 'Year':'year', 'Quarter':'quarter',
                                                          'Product Name':'product_name', 'Units Reimbursed':'units_reimbursed',
                                                          'Suppression Used':'med_data_suppresion', 
                                                          'Number of Prescriptions':'no_of_prescriptions',
                                                          'Total Amount Reimbursed':'total_amt_reimbursed', 
                                                          'Medicaid Amount Reimbursed':'medicaid_amt_reimbursed',
                                                          'Non Medicaid Amount Reimbursed':'non_medicaid_amt_reimbursed', 
                                                          'Quarter Begin':'quarter_begin', 
                                                          'Quarter Begin Date':'quarter_begin_date','Latitude':'latitude', 
                                                          'Longitude':'longitude', 
                                                          'Location':'location', 'NDC':'national_drug_code'})
    
    
        
cleaned_state_dud_df.head()
```
